```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
import os
from itables import init_notebook_mode
init_notebook_mode(all_interactive=True)
```

```python
df = pd.read_csv('DATA/df_Normalized.csv')
df_in = pd.read_csv('DATA/data_test_users.csv')
```

```python
ohe = pd.get_dummies(df.intent)

df['1'] = ohe['1'] + ohe['1,2'] + ohe['1,3'] + ohe['1,4'] + ohe['1,5']
df['2'] = ohe['2'] + ohe['1,2']
df['3'] = ohe['3'] + ohe['1,3'] 
df['4'] = ohe['4'] + ohe['1,4'] 
df['5'] = ohe['5'] + ohe['1,5'] 
```

```python
def seg2label(P):      
    PR = []
    P = np.asarray(P)
    for p in P:
        if (p>0.95).any():
            p = p>(p.max()-0.05)
            p = df.columns[-5:].values[p]
            p = ",".join(p)
        else:
            p = df.columns[-5:].values[np.argmax(p)]
        
        PR.append(p)
        
    return PR
```

```python
preds = []
for f in os.listdir():
    if f.endswith('.xlsx'):
        print(f)
        if f.endswith('w.xlsx'):
            preds.append(pd.read_excel(f, index_col=0))
        preds.append(pd.read_excel(f, index_col=0))
```

```python
preds[1]
```

# Mean

```python
df_inf = df_in.copy()

P = []
for p in preds:
    print(p.shape)
    p = p.iloc[:,-5:]
    P.append(p)

P = np.asarray(P).mean(axis=0)

df_inf['intent'] = seg2label(P)



df_inf
```
